```python
import random
def choose_word():
words=["apple","banana","cherry","mango","kiwi","orange","watermelon","grapes","lemon","avacoda","strawberry","dragonfruit","papaya","pineapple","pomegranate"]
    return random.choice(words)
def display_word(word,guessed_letters):
    display=""
    for letter in word:
        if letter in guessed_letters:
            display+=letter
        else:
            display+="_"
    return display
name=input("Enter your name:")
print("Welcome to Hangman",name)
print('''Rules for the game are:
1)This is a fruit guessing game,so guess for a fruit.
2)A row of dashes represents the word to guess.
3)If the player guesses a letter in the word,the script writes it in all its correct positions.
4)The player has 6 turns to guess the word.
5)If the player finds the word before 6 turns,he/she wins the game.''')
def hangman():
    word=choose_word()
    guessed_letters=[]
    incorrect_guesses=0
    max_incorrect_guesses=6
    while True:
        print("\nWord to guess:",display_word(word,guessed_letters))
        guess=input("Guess a letter:").lower()
        if guess in guessed_letters:
            print("You already guessed that letter.Try again.")
            continue
        guessed_letters.append(guess)
        if guess not in word:
            incorrect_guesses+=1
            print("Incorrect guess.You have",max_incorrect_guesses-incorrect_guesses,"attempts left.")
        if incorrect_guesses==max_incorrect_guesses:
            print("\nYou have run out of attempts.The word was:",word)
            break
        if all(letter in guessed_letters for letter in word):
            print("\nCongrulations!! You have guessed the word:",word)
            break
while True:
    hangman()
    play_again=input("Do you want to play again?(yes/no):").lower()
    if play_again!='yes':
        print("Thanks for playing.GoodBye!")
        break
```
